```python
def lexical_divercity(text):
    return len(set(text)) / len(text)
```

```python
def percentage(word, text):
    return text.count(word) / len(text) * 100
```

```python
percentage('death', book.text3)
```

# 3   Computing with Language: Simple Statistics

## 3.1   Frequency Distributions

How can we automatically identify the words of a text that are most informative about the topic and genre of the text? Imagine how you might go about finding the 50 most frequent words of a book. 

```python
fdict1 = nltk.FreqDist(book.text3)
```

```python
fdict1.most_common(50)
```

```python
plt.figure(figsize=(15, 8))
fdict1.plot(50, cumulative=True);
```

If the frequent words don't help us, how about the words that occur once only, the so-called hapaxes?

```python
# the words that occur once only
fdict1.hapaxes()
```

## 3.2   Fine-grained Selection of Words

```python
len(book.text1)
```

```python
long_words = [w for w in V if len(w) > 15]
len(long_words)
```

#  Task -> find words that characterize a text

```python
# longer than seven characters, that occur more than seven times
[(w,fdict1[w]) for w in V if len(w) > 7 and fdict1[w] > 7 and fdict1[w] < 17]
```
